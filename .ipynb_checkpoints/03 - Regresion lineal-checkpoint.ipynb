```python
import numpy as np
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

```python
data = sns.load_dataset('iris')
data.head(10)
```

```python
data['species'].unique()
```

```python
data['species'].value_counts()
```

```python
x = data['petal_length']
y = data['petal_width']
```

```python
# Agregar columna constante = 1 para calculo del intercepto
x = sm.add_constant(x)
```

```python
# apliar el modelo de minimo ordinarios cuadrados
resutlado = sm.OLS(y,x).fit()
```

```python
resutlado.summary()
```

# r de pearson 
* -1 = negativo(total);
*  0 = no relacionado;
*  1 = Positiva(total)

```python
resutlado.params
```

```python
resutlado.rsquared
```

```python
import math
```

```python
math.sqrt(resutlado.rsquared)
```

# Grafico de regresion

```python
sns.set_theme(color_codes=True)
```

```python
ax = sns.regplot(data=data, x='petal_length', y='petal_width')
ax.set_xlabel('longitud de petalo')
ax.set_ylabel("ancho de petalo")
ax.set_title('relacion entre la longitud y el ancho del petalo')
ax.plot();
```

```python
ax2 = sns.jointplot(data=data, x='petal_length', y='petal_width', kind='reg', truncate=False, color='m', height=7)
```

```python
# diagramas de dispersion de los atributes emparejados
sns.pairplot(data, hue='species', height=2, palette='colorblind');
```
