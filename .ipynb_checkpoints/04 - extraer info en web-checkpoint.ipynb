# Web Scraping

Se requiere el modulo de BeautifulSpoup 

* pip install bs4


```python
import requests
from bs4 import BeautifulSoup
```

```python
# Enlace a la pagina 
url = 'https://es.wikipedia.org/wiki/Organizaci%C3%B3n_territorial_de_Guatemala'
```

```python
response = requests.get(url)
```

```python
print('Status code',url,':', response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tabla = soup.find('table', {'class':'wikitable'})
```

```python
import pandas as pd
```

```python
tabla = pd.read_html(str(tabla))
tabla
```

```python
departamento = tabla[1]
```

```python
departamento.columns = ['Departamento', 'Cabecera', 'Superficie', 'Poblacion']
departamento
```

```python
region = soup.find_all('table')[2]
region = pd.read_html(str(region))[0]
region.columns['Region', 'Departamentos', 'Extension', 'Poblacion', 'Ciudad Mas Poblada']
region
```

```python
segeplan = soup.find_all('table')[3]
segeplan = pd.read_html(str(segeplan))[0]
segeplan.columns['Region', 'Departamento', 'Municipios']
segeplan
```

```python
subregion = soup.find_all('table')[4]
subregion = pd.read_html(str(subregion))[0]
subregion.columns['Subregion','Departamento','Municipios']
subregion
```

```python
departamento['Superficie'] = list(int(x.replace('\xa0','')) for x in departamento['Superficie'])
departamento
```
