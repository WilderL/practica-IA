```python
#importar librelias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
# cargar el dataset
data = sns.load_dataset('iris')
# Separacion de datos en entrenamiento de clase y atributos
train, test = train_test_split(data, test_size=0.4,stratify=data['species'], random_state=19)
```

```python
# separacion de la etiqueta de clase/variable de respuesta  <Y> y atributos <X>
# Conjunto de entrenamiento
X_train = train[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y_train = train['species']
# Conjunto de pruebas
X_test = test[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y_test = test['species']
```

```python
# Clasificador Gaussiano Naive Bayes
modelo_gnb =GaussianNB()
y_pred = modelo_gnb.fit(X_train, Y_train).predict(X_test)
print('Lapredicion del clasificador Gaussiano Naive Bayes en los datos de prueba es: {:0.3f}'.format(metrics.accuracy_score(y_pred, Y_test)))
```

```python
pred2 = ['petal_length', 'petal_width']
mod_gnb_2pred = GaussianNB()
y_pred2 = mod_gnb_2pred.fit(X_train[pred2], Y_train).predict(X_test[pred2])
print('Lapredicion del clasificador Gaussiano Naive Bayes en los datos de prueba es: {:0.3f}'.format(metrics.accuracy_score(y_pred2, Y_test)))
```

```python
test['Prediccion'] = y_pred
test['Prediccion 2'] = y_pred2
test
```
