## Introduccion a OpenCV

```python
# Importar paquetes
from matplotlib import pyplot as plt
import numpy as np
import imageio
from skimage import color

# biblioteca Open CV
import cv2
import common
# !pip install common

# Para control de dimensiones de la imagen
import pylab
```

```python
# Modificar las dimensoines de visualizacion
pylab.rcParams['figure.figsize'] = (6.0, 4.0)
ruta = 'wall-e.jpg'
input_image = cv2.imread(ruta)
```

```python
plt.imshow(input_image);
```

```python
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB));
```

```python
# Inversion vertical de la imagen (0)
flippled_img = cv2.flip(input_image, 0)
plt.imshow(flippled_img);
```

```python
# Inversion hotizontal de la imagen (1)
flippled_img_1 = cv2.flip(input_image, 1)
plt.imshow(flippled_img_1);
```

```python
# Imagen traspuesta
transpuesta = cv2.transpose(input_image)
;plt.imshow(transpuesta);
```

```python
plt.imshow(cv2.cvtColor(cv2.flip(transpuesta,1), cv2.COLOR_BGR2RGB));
```

## Operaciones aritmetricas con OpenCV

```python
# Crear una matriz de h,w,3
blank_image = np.zeros((input_image.shape), np.uint8)
# Crear el marco
blank_image[125:255, 350:450, 1] = 50
```

```python
plt.imshow(blank_image);
```

```python
nueva = cv2.cvtColor(cv2.add(blank_image, input_image), cv2.COLOR_BGR2RGB)
plt.imshow(nueva);
```

```python
# Filtro de desenfoque
d = 5
img_desenfoque = cv2.GaussianBlur(input_image, (2*d+1, 2*d+1), -1)
plt.imshow(cv2.cvtColor(img_desenfoque, cv2.COLOR_BGR2RGB));
```

```python
# Filtro de desenfoque
d = 5
img_desenfoque = cv2.GaussianBlur(input_image, (2*d+1, 2*d+1), -1)[d:-3, d:-d]
plt.imshow(cv2.cvtColor(img_desenfoque, cv2.COLOR_BGR2RGB));
```

```python
grayImage = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImage, cmap='gray');
```

```python
sobelX = cv2.Sobel(grayImage, cv2.CV_64F, 1, 0, ksize=9)
plt.imshow(sobelX, cmap='gray');
```

```python
sobelY = cv2.Sobel(grayImage, cv2.CV_64F, 0, 1, ksize=9)
plt.imshow(sobelY, cmap='gray');
```

```python
sobelXY = cv2.Sobel(grayImage, cv2.CV_64F, 1, 1, ksize=9)
plt.imshow(sobelXY, cmap='gray');
```

## Maximos y Minimos

```python
for i in range(0,3):
    minimo, maximo, min_loc, max_loc = cv2.minMaxLoc(input_image[:,:,i])
    print('min {} esta ubicadoen {} y el max {} esta ubicado en {}'.format(minimo, min_loc, maximo, max_loc))
```
