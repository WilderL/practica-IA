**Ejemplo de graficas en una presentacion interactiva**

$$\text{Sea } 0 \leqslant x \leqslant 10$$
$$f_1(X) = x$$
$$f_2(X) = x^2$$
$$f_3(X) = x^3$$

````python
import matplotlib.pyplot as plf
import numpy as np
````

````python
# Rango de valores para x
# np.arange(inicio,fin,paso)
x = np.arange(0,10,0.4)
````

````python
x
````

````python
# graficar unan funcion lineal (x,y,formato)
plf.plot(x,x,'g.')
````

````python
y1=x
y2=x**2
y3=x**3
````

````python
plf.plot(x,y1,'g--', x,y2,'bs', x,y3,'r^')
plf.show()
````
